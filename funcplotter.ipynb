```python
import numpy as np
import plotly.graph_objs as go
import plotly.io as pio
```

```python
# Define the function
def f(x,y):
    return np.exp(x)+ np.exp(y) + x**2 + y**2
```

```python
# Create grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create interactive 3D surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')])
fig.update_layout(
    title='f(x, y) = log(1 + x + y)',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='f(x, y)'
    ),
    width=800,
    height=700
)

# Show the plot
pio.show(fig)
```
