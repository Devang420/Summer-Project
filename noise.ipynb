```python
import numpy as np
import plotly.graph_objects as go

# Grid
L = 2 * np.pi
N = 200
x = np.linspace(-1*L, L, N)
y = np.linspace(-1*L, L, N)
X, Y = np.meshgrid(x, y)

# Base function
Z = np.sin(X) + Y**2

# Add noise component
K = 10  
for k1 in range(1, K+1):
    for k2 in range(1, K+1):
        Z += 0.09*np.cos(k1 * X + k2 * Y)

# Interactive 3D surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')])
fig.update_layout(
    title='f(x, y) = sin(x) + y + structured noise',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='f(x, y)',
        aspectratio=dict(x=1, y=1, z=0.5)
    ),
    width=900,
    height=700
)
fig.show()
```
